```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize variables
sample_sizes = range(500, 10500, 500)  
e = 0.01 

prob_values = [] 
prob1_values = []

for n in sample_sizes:
    count = 0
    count1 = 0
    
    for i in range(10000):
        x = np.random.uniform(1, 4, n)
        
        
        if np.abs(np.mean(x) - 2.5) > e:
            count += 1
        if np.abs(np.var(x) - 0.75) > e:
            count1 += 1

    # Calculate probabilities
    prob = count / 10000
    prob1 = count1 / 10000

    # Store results
    prob_values.append(prob)
    prob1_values.append(prob1)

# Plot results
plt.plot(sample_sizes, prob_values, color='red', ls='-', label="Mean deviation")
plt.plot(sample_sizes, prob1_values, color='green', ls='-', label="Variance deviation")
plt.xlabel("Sample size")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

mean =[]
variance= []
k_4 = ((4-1)**4)/80  # fourth order central moment
for i in range(10000):
    x= np.random.uniform(1,4,100)
    mean.append(np.mean(x))
    variance.append(np.var(x,ddof=1))

mean= np.array(mean)
variance= np.array(variance)  # we are converting to array so that it is easier to do mathematical operations without loops

x_vals= np.linspace(-4,4,500)
dist_1 = (np.sqrt(100)*(mean- 2.5))/np.sqrt(0.75)
dist_2 = (np.sqrt(100)*(variance- 0.75))/ np.sqrt(k_4 - 0.75**2)

fig,axes= plt.subplots(1,2, figsize= (13,5))
axes[0].hist(dist_1, bins=30, density= True, color='blue')
axes[0].plot(x_vals, norm.pdf(x_vals, 0,1), color= 'red')
axes[0].set_title('CLT for mean')

axes[1].hist(dist_2, bins=30, density= True, color='green')
axes[1].plot(x_vals, norm.pdf(x_vals, 0,1), color= 'red')
axes[1].set_title('CLT for variance')
plt.show()
```
